5 Techniques of Sampling Imbalanced Data

```python
import pandas as pd
import numpy as np
import imblearn
```

```python
#Divide the dataset to 
df= pd.read_csv('Creditcard_data.csv')
#x contains all the features
x= df.drop('Class',axis=1)
#y contains the target
y = df['Class']
```

                                    SAMPLING I 

Random Under-Sampling with imblearn

```python
from imblearn.under_sampling import RandomUnderSampler

rus= RandomUnderSampler(random_state=42, replacement=True)
#x is the predictor, and y is the target variable
x_rus , y_rus = rus.fit_resample(x,y) 
```

```python
x_rus
```

```python
y_rus
```

                                SAMPLING II

Random Over-Sampling With imblearn

```python
from imblearn.over_sampling import RandomOverSampler

ros= RandomOverSampler(random_state=42)
x_ros, y_ros = ros.fit_resample(x,y)
```

```python
x_ros
```

```python
y_ros
```

                                SAMPLING III

Under-Sampling: Tomek Links

```python
from imblearn.under_sampling import TomekLinks
from collections import Counter

tl= TomekLinks(sampling_strategy='majority')

x_tl , y_tl = tl.fit_resample(x,y)
print('Original dataset shape:', Counter(y))
print('Resampled dataset shape:', Counter(y_tl))
```

                                SAMPLING IV

Synthetic Minority Oversampling Technique (SMOTE)

```python
from imblearn.over_sampling import SMOTE

smote= SMOTE()

x_smote , y_smote = smote.fit_resample(x,y)

print("Original dataset shape: ", Counter(y))
print("Resample dataset shape: ", Counter(y_smote))
```

Smote has created 763 rows from 9 rows by using K-nearest neighbour algorithm on the rows classified as 1

                                SAMPLING V

NearMiss

```python
from imblearn.under_sampling import NearMiss

nm = NearMiss()

x_nm, y_nm = nm.fit_resample(x,y)
print("Original Dataset Shape: ", Counter(y))
print("Resample Dataset Shape: ", Counter(y_nm))
```

Applying 5 ML Models on these Samples to find Accuracy

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, f1_score
import warnings
warnings.filterwarnings('ignore')

models = {
    'Logistic Regression': LogisticRegression(max_iter=1000),
    'Decision Tree': DecisionTreeClassifier(random_state=42),
    'Random Forest': RandomForestClassifier(random_state=42),
    'SVM': SVC(random_state=42),
    'KNN': KNeighborsClassifier()
}

datasets = {
    'Random Under-Sampling': (x_rus, y_rus),
    'Random Over-Sampling': (x_ros, y_ros),
    'Tomek Links': (x_tl, y_tl),
    'SMOTE': (x_smote,y_smote),
    'NearMiss': (x_nm,y_nm)
}

results=[]

#Iterate over the different Sampling datasets to train all models
for sampling_name, (X_sampled, y_sampled) in datasets.items():
    X_train, X_test, y_train, y_test= train_test_split(X_sampled, y_sampled, test_size=0.3, random_state=42)
    #Apply models on each sample
    for model_name, model in models.items():
        #Train the model
        model.fit(X_train, y_train)
        #Predict
        y_pred= model.predict(X_test)
        #Calculate accuracy
        accuracy = accuracy_score(y_test,y_pred)
        f1= f1_score(y_test, y_pred)
        results.append({
            'Sampling Method': sampling_name,
            'Model':model_name,
            'Accuracy': round(accuracy, 4),
            'F1-Score': round(f1, 4)
        })

results_df= pd.DataFrame(results)
print(results_df)
results_df.to_csv('output.csv')
```

```python
pivot_table = results_df.pivot(index='Model', columns='Sampling Method', values='Accuracy')

# Reorder columns if needed
pivot_table = pivot_table[['Random Under-Sampling', 'Random Over-Sampling', 'Tomek Links', 'SMOTE', 'NearMiss']]

print(pivot_table)

# Save to CSV
pivot_table.to_csv('accuracy_comparison.csv')
```
